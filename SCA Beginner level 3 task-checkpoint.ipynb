```python
import matplotlib.pyplot as plt
```

```python
#MATPLOTLIB
```

```python
CompanySales = pd.read_csv("company_sales_data.xls")
CompanySales
```

```python
plt.figure(figsize = (10,8))
plt.plot(CompanySales['month_number'],
           CompanySales['total_profit'])
plt.xlabel('Month of the Year')
plt.ylabel('Total Profit')
plt.title('Company Profit per Month')
plt.xticks(ticks = range(1,13))
plt.yticks(ticks = range(100000,600000,100000))
plt.show()
```

```python
labels = ['FaceCream', 'FaceWash', 'Toothpaste', 'BathingSoap', 'Shampoo', 'Moisturizer' ]
plt.figure(figsize = (10,8))
plt.plot(CompanySales['month_number'], CompanySales['facecream'], marker = "o")
plt.plot(CompanySales['month_number'], CompanySales['facewash'], marker = "o")
plt.plot(CompanySales['month_number'], CompanySales['toothpaste'], marker = "o")
plt.plot(CompanySales['month_number'], CompanySales['bathingsoap'], marker = "o")
plt.plot(CompanySales['month_number'], CompanySales['shampoo'], marker = "o")
plt.plot(CompanySales['month_number'], CompanySales['moisturizer'],marker = "o")
plt.xlabel('Month of the Year')
plt.ylabel('Sales units in numbers')
plt.title('Sales data per products')
plt.xticks(ticks = range(1,13))
plt.yticks(ticks = range(1000,11000,1000))
plt.legend(labels = labels, fontsize = 'large')
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.scatter(CompanySales['month_number'],
           CompanySales['toothpaste'])
plt.xlabel('Month of the Year')
plt.ylabel('Number of Units Sold')
plt.title('Toothpaste Sales per Month')
plt.xticks(ticks = range(1,13))
plt.yticks(ticks = range(4500,9000,500))
plt.grid(True, linestyle = '--')
plt.show()
```

```python
plt.figure(figsize = (10,8))
x = np.arange(1, 13)
plt.bar(CompanySales['month_number'] , CompanySales['facecream'], 0.4,  label = 'facecream')
plt.bar(CompanySales['month_number'] + 0.4, CompanySales['facewash'], 0.4, label = 'facewash')
plt.xlabel('Month of the Year')
plt.ylabel('Sales units in numbers')
plt.title('Facewash and Facecream Sales data')
plt.xticks(ticks = range(1,13))
plt.yticks(ticks = range(0,4000,500))
plt.legend(loc = 2, labels = labels, fontsize = 'large')
plt.grid(True, linestyle = '--')
plt.show()
```

```python
plt.figure(figsize =(10,8))
n = []
for i in CompanySales.columns: n.append(CompanySales.loc[:,i].sum())
Products = ['FaceCream', 'FaceWash', 'Toothpaste', 'BathingSoap', 'Shampoo', 'Moisturizer' ]
plt.pie(n[1:7], labels = Products, autopct = '%.1f%%')
plt.title("Sales Data")
plt.legend()
plt.show()
```
